```python
print(msno.matrix(all_players.sort_values(by='ACTIVE_FROM')))
print(all_players.isna().sum()/len(all_players))
```

Los COLLEGE vacíos se deben a jugadores que no han ido al COLLEGE. Han ido directamente a la NBA sin pasar por el college (Kevin Garnett) o han llegado desde Europa (Álex Abrines). Los rellenamos con 'No College'.

```python
all_players[(all_players['NAME']=='Kevin Garnett') | (all_players['NAME'] == 'Álex Abrines')]
```

```python
all_players_1 = all_players.copy()
all_players_1['COLLEGE'] = all_players['COLLEGE'].fillna('No College')
```

Los que tienen 'DRAFT_YEAR', 'DRAFT_PICK' y 'DRAFT_TEAM' sin datos es no han sido seleccionados en el draft. Vemos el caso de Bill Edwards que le aparece 'DRAFT_PICK', pero no el año ni el equipo:

```python
all_players_1[all_players_1['DRAFT_PICK'].isna() & all_players_1['DRAFT_YEAR']>0]
```

Observamos que el DRAFT_YEAR es 1953, mientras que ACTIVE_FROM es 1994. Con una búsqueda en wikipedia (https://en.wikipedia.org/wiki/Bill_Edwards_(basketball)) vemos que no fue seleccionado en el draft. Los señalamos como "undrafted".

```python
all_players_2 = all_players_1.copy()

all_players_2['DRAFT_PICK'] = all_players_2['DRAFT_PICK'].fillna('Undrafted')
all_players_2['DRAFT_TEAM'] = all_players_2['DRAFT_TEAM'].fillna('Undrafted')
all_players_2['DRAFT_YEAR'] = all_players_2['DRAFT_YEAR'].fillna('Undrafted')
all_players_2.loc[714, 'DRAFT_YEAR'] = 'Undrafted'
```

Analizamos los tipos de variables. Todo correcto excepto el HEIGHT.

```python
all_players_2.dtypes
```

Modificamos 'BIRTH_DATE', 'HEIGHT' y 'WEIGHT' para que tengan unidades de medida más manejables y comprobamos que todos los tipos de datos sean correctos

```python
all_players_3 = all_players_2.copy()
all_players_3['BIRTH_DATE(year)'] = all_players_3['BIRTH_DATE'].dt.strftime('%Y')
all_players_3['BIRTH_DATE(year)'] = all_players_3['BIRTH_DATE(year)'].apply(lambda x: int(x))
all_players_3['HEIGHT(cm)'] = all_players_3['HEIGHT'].apply(lambda x: float(x[:1])*30.48 + float(x[2:])*2.54 )
all_players_3['WEIGHT(kg)'] = all_players_3['WEIGHT'].apply(lambda x: round(x/2.204622,2))
all_players_3 = all_players_3.drop(['BIRTH_DATE', 'HEIGHT', 'WEIGHT'], 1)
```

```python
all_players_3.dtypes
```

Guardamos el archivo limpio

```python
all_players_3 = all_players_3.drop_duplicates()
all_players_3  = all_players_3.reset_index()
```

```python
all_players_3 = all_players_3.drop('index', 1)
```

```python
all_players_3.to_csv('Clean_data/All_players_Clean.csv')
all_players_3.to_excel('Clean_data/All_players_Clean.xlsx')
```
